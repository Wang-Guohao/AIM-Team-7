```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.externals import joblib
import pandas as pd
import numpy as np
import ssl

df = pd.read_csv('./Tumor Information_train.csv')
dftest = pd.read_csv('./Tumor Information_test.csv')

df_backgroud = pd.read_csv('./Background and Helpful Info.csv')
df_beverage = pd.read_csv('./Beverages.csv')
df_current_lifestyle = pd.read_csv('./Current Lifestyle.csv')
df_exercise = pd.read_csv('./Exercise.csv')
df_family_history = pd.read_csv('./Family History.csv')
df_food_habits = pd.read_csv('./Food Habits.csv')
df_foods = pd.read_csv('./Foods.csv')
df_herbals = pd.read_csv('./Herbals.csv')
df_lifetime_supplements = pd.read_csv('./Lifetime use of Supplements.csv')
df_medical_history = pd.read_csv('./MedicalHistory.csv')
df_medication_use_history = pd.read_csv('./MedicationUseHistory.csv')
df_multivitamins = pd.read_csv('./Multivitamins.csv')
df_nutrient = pd.read_csv('Nutrient Calc.csv')
df_smoking = pd.read_csv('./Smoking History.csv')
df_vitamins = pd.read_csv('./Vitamins Minerals.csv')
df_womens_health = pd.read_csv('./WomensHealth.csv')


df.columns = [col.lower() for col in df.columns]
dftest.columns = [col.lower() for col in dftest.columns]

df_backgroud.columns = [col.lower() for col in df_backgroud.columns]
df_beverage.columns = [col.lower() for col in df_beverage.columns]
df_current_lifestyle.columns = [col.lower() for col in df_current_lifestyle.columns]
df_exercise.columns = [col.lower() for col in df_exercise.columns]
df_family_history.columns = [col.lower() for col in df_family_history.columns]
df_food_habits.columns = [col.lower() for col in df_food_habits.columns]
df_foods.columns = [col.lower() for col in df_foods.columns]
df_herbals.columns = [col.lower() for col in df_herbals.columns]
df_lifetime_supplements.columns = [col.lower() for col in df_lifetime_supplements.columns]
df_medical_history.columns = [col.lower() for col in df_medical_history.columns]
df_medication_use_history.columns = [col.lower() for col in df_medication_use_history.columns]
df_multivitamins.columns = [col.lower() for col in df_multivitamins.columns]
df_nutrient.columns = [col.lower() for col in df_nutrient.columns]
df_smoking.columns = [col.lower() for col in df_smoking.columns]
df_vitamins.columns = [col.lower() for col in df_vitamins.columns]
df_womens_health.columns = [col.lower() for col in df_womens_health.columns]

df_training = df.merge(df_backgroud, on='qbarcode', how='left')\
.merge(df_beverage, on='qbarcode', how='left')\
.merge(df_current_lifestyle, on='qbarcode', how='left')\
.merge(df_exercise, on='qbarcode', how='left')\
.merge(df_family_history, on='qbarcode', how='left')\
.merge(df_food_habits, on='qbarcode', how='left')\
.merge(df_foods, on='qbarcode', how='left')\
.merge(df_herbals, on='qbarcode', how='left')\
.merge(df_lifetime_supplements, on='qbarcode', how='left')\
.merge(df_medical_history, on='qbarcode', how='left')\
.merge(df_medication_use_history, on='qbarcode', how='left')\
.merge(df_multivitamins, on='qbarcode', how='left')\
.merge(df_nutrient,on='qbarcode',how='left')\
.merge(df_smoking, on='qbarcode', how='left')\
.merge(df_vitamins, on='qbarcode', how='left')\
.merge(df_womens_health, on='qbarcode', how='left')

df_test = dftest.merge(df_backgroud, on='qbarcode', how='left')\
.merge(df_beverage, on='qbarcode', how='left')\
.merge(df_current_lifestyle, on='qbarcode', how='left')\
.merge(df_exercise, on='qbarcode', how='left')\
.merge(df_family_history, on='qbarcode', how='left')\
.merge(df_food_habits, on='qbarcode', how='left')\
.merge(df_foods, on='qbarcode', how='left')\
.merge(df_herbals, on='qbarcode', how='left')\
.merge(df_lifetime_supplements, on='qbarcode', how='left')\
.merge(df_medical_history, on='qbarcode', how='left')\
.merge(df_medication_use_history, on='qbarcode', how='left')\
.merge(df_multivitamins, on='qbarcode', how='left')\
.merge(df_nutrient,on='qbarcode',how='left')\
.merge(df_smoking, on='qbarcode', how='left')\
.merge(df_vitamins, on='qbarcode', how='left')\
.merge(df_womens_health, on='qbarcode', how='left')
```

```python
df_training.columns = [col.lower() for col in df_training.columns]
df_test.columns = [col.lower() for col in df_test.columns]

print(df_training.columns)
print(df_test.columns,'\n')
```

```python
df_all_train = pd.read_csv('./allTrain.csv')
df_all_train.columns = [col.lower() for col in df_all_train.columns]
print(df_all_train.columns)
print(len(df_all_train.columns))
```

```python
all_data = pd.concat([df_training,df_test], sort = True)
cols = [col for col in all_data.columns if col not in ['patientstatus', 'qbarcode']]

types = all_data.dtypes
cat_columns = [t[0] for t in types.iteritems() if ((t[1] not in ['int64', 'float64']))]
print('Label encoding categorical columns:', cat_columns)
print(len(cols))
```

```python
from sklearn import preprocessing
from sklearn.model_selection import KFold, StratifiedKFold, GroupKFold
lbl = preprocessing.LabelEncoder()
for col in cat_columns:
    all_data[col] = lbl.fit_transform(all_data[col].astype(str))
train_processed_data = all_data.iloc[:len(df_training)]
test_processed_data = all_data.iloc[len(df_training):]

train_processed_data.fillna(0, inplace=True)
test_processed_data.fillna(0, inplace=True)

train_processed_data.drop(columns=['litho1', 'litho2'], inplace=True)
test_processed_data.drop(columns=['litho1', 'litho2'], inplace=True)
```

```python
cols = [col for col in train_processed_data.columns if col not in ['patientstatus', 'qbarcode']]
X_train, X_val, y_train, y_val = train_test_split(
    train_processed_data[cols], train_processed_data['patientstatus'], test_size=0.2, random_state=65)

X_train.columns

clf = LogisticRegression(random_state=2).fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score
out = clf.predict(X_val)
accuracy_score(y_val, out)
```
